Matrices and Vectors: an overview

Vector: A vector is a matrix with either only one column or only one row. A column vector has only one column. A row vector has only one row.


Matrix: A matrix is two-dimensional array of numbers.

Dimentionality: The dimension of a matrix is expressed as number of rows × number of columns. For instance, a matrix with 10 rows and 3 columns is said to be a 10 × 3 matrix. The vectors written in Example 1 above are 3 × 1 matrices

```python
# show how to represent vectors and matrices in Python

import numpy as np

#create vector
a = np.array([10,20,30])

#all zeros 
zero_vector=np.zeros(5)
print("zero_vector",zero_vector)

#from 1 to 10
vector_1_10=np.arange(10)
print("vector_1_10: ",vector_1_10)

#change position 7 in the vector (careful, Python is 0 based!)
vector_1_10[6]= 5
print("vector_1_10: ",vector_1_10)

#create matrix

# all 0s
matrix_all_zeros= np.zeros((2, 3))
print("matrix_all_zeros: ",matrix_all_zeros)

# with filled columns
matrix=np.matrix([[10,20],[5,10],[100,10]])
print("matrix: ",matrix)

# random of fixed size
matrix_random= np.random.random((3, 3))
print("matrix_random: ",matrix_random)
# change the element in the 3rd row and 2nd column
matrix[2,1]=5
print("matrix: ",matrix)

#print only the first row
print("First row of matrix: ",matrix[0,:])

#print only the first row
print("First column of matrix: ",matrix[:,0])
```

Matrix Transformation


Mattrix $A$ is represented as $\mathbf{A}^\intercal$ to denote the transpose of $A$

If matrix $A$ =   \begin{bmatrix}
    1 & 2   \\
    3 & 4 
  \end{bmatrix} Then, 

$\mathbf{A}^\intercal$= \begin{bmatrix}
    1 & 3   \\
    2 & 4 
  \end{bmatrix}
  
  Matrix $A$ is represented as $\mathbf{A}^{-1}$ to represent the inverse of $A$. 
  
If matrix $A$ =   \begin{bmatrix}
    a & b   \\
    c & d 
  \end{bmatrix}  Then, 
  
  $\mathbf{A}^{-1}$ = 1/det *\begin{bmatrix}
    d & -b   \\
    -c & a 
  \end{bmatrix} where det=ab-bc



```python
# Matrix Transpose
transposed_matrix = matrix.transpose()
print(matrix)
print(transposed_matrix)

# Matrix Inverse
from numpy.linalg import inv
#matrix needs to be square to invert
invertable_matrix=matrix_random[0:2,0:2]
print(invertable_matrix)

inversed_matrix=inv(invertable_matrix)
print(inversed_matrix)
```

Matrix Combinations

Two matrices may be added if and only if they have the same dimensions (same number of rows and also same number of columns as each other. To add two matrices, add corresponding elements (in terms of location).

Let $A$ =  \begin{bmatrix}
    a & b   \\
    c & d 
  \end{bmatrix} and let $B$ =  \begin{bmatrix}
    e & f   \\
    g & h 
  \end{bmatrix}  
  then $A + B$ =  \begin{bmatrix}
    a+e & b+f   \\
    c+g & d+h 
  \end{bmatrix}   
  
  Note:  matrix is subtracted from another in the same way that matrices are added. To do the subtraction, subtract each element of the first matrix from the corresponding (in location) element of the first matrix.
  
  

  
  
  



```python
# vec and vec addition

#method 1
new_Vec_1= a+vector_1_10[0:3]

#method 2
new_Vec_2= np.add(a,vector_1_10[0:3])
print(new_Vec_1,new_Vec_2)

# matrix and matrix addition (make sure dimensions match!)
new_matrix = inversed_matrix+matrix[0:2,:]
print(new_matrix)
```

Matrix Multiplication

Matrix are multiplied in two different ways:
  
  1- The word scalar is a synonym for a numerical constant. (In matrix terms, a scalar is a matrix with one row and one column.)

To multiply a matrix by a scalar, multiply each element in the matrix by the scalar.

Let $x$ = 5 and $A$ = \begin{bmatrix}
    a & b   \\
    c & d 
  \end{bmatrix} hence, $x . A$ = 5.\begin{bmatrix}
    a & b   \\
    c & d 
  \end{bmatrix} = \begin{bmatrix}
    5a & 5b   \\
    5c & 5d 
  \end{bmatrix}
  

2- To be able to perform the matrix multiplication $C = A . B$ , the number of columns in the matrix 
$A$  must equal the number of rows in the matrix $B$.The element in the ith row and jth column of the answer matrix  $C$, is the cross product sum of the ith row of the matrix  $A$ and the jth column of the matrix $B$. This is done of all combinations of rows of 
$A$ and columns of $B$.

Let $A$ = \begin{bmatrix}
    1 & 2   \\
    3 & 4 
  \end{bmatrix} and let 

$B$= \begin{bmatrix}
    1 & 3   \\
    2 & 4 
  \end{bmatrix} then,
  
  $A . B$ = \begin{bmatrix}
    [1*1+2*2] & [1*3+2*4]   \\
   [3*1+4*2] & [3*3+4*4] 
  \end{bmatrix} = \begin{bmatrix}
    5 & 11   \\
    11 & 25 
  \end{bmatrix}
  
  NOTE: $A * B$ = $\neq$ $B * A$
  
  



```python
# scalar and vector

scaled_vector= 5 * new_Vec_1

print(new_Vec_1,scaled_vector)

#scalar and matrix

scaled_matrix= 5*matrix

print(matrix, scaled_matrix)

# vec and matrix multiplication (dimensions need to match!) 

#Note new_vec_1 dimensions 1 x 3
#Note matrix dimensions: 3 x 2
multiplied_matrix= new_Vec_1 * matrix
#Note multiplied_matrix dimensions:  1 x 2
print(multiplied_matrix)

# vec x vec -> scalar
vector_mutiplication_to_scalar = new_Vec_1 * new_Vec_2.transpose()
print(vector_mutiplication_to_scalar)

# vec x vec -> matrix
vector_mutiplication_to_matrix = new_Vec_1.transpose() * new_Vec_2
print(vector_mutiplication_to_matrix)

# matrix x matrix
matrix_multiplication = matrix_random *matrix
print(matrix_multiplication)

```

Extracting innitial information from a dataset, 


Population Mean :
$\mu$ = $E[Y]$

Sample Mean : $\bar{x} = \frac1{n}\sum_{i=1}^n{x_i}$

This is equivalent to the following 

population mean : $\min_{\mu}E[(Y-\mu)^2]$ => $\mu$ = $E[Y]$

sample mean : $\min_\bar{x} = \frac1{n}\sum_{i=1}^n{(x_i-\mu)}^2$ => $\bar{x} = \frac1{n}\sum_{i=1}^n{x_i}$

Variance: 
$\sigma^2 = \frac{\displaystyle\sum_{i=1}^{n}(x_i - \mu)^2} {n}$

Preprocessing for algorithm:

Its a very common practice in Machine Learning to split the data into:

Training : the subset of data that is used to train the moel (algorithm) [1/3]
Validation : the sub subset of data that is used to validate the performace of the algorithm [2/3]
Test: The data set that is used to test the the algorithm [3/3]

The above aproches are performed after data randomisation which allows to get ride of any patterns/bias that exist within the dataset.

PART 3:  The first predictive algorithm: Linear Regression


Linear regression is a statistical method that is used mostly for 2 reasons:
1- how much of the variation in the dependent varaible/ target/ regressand/ output is explained d data/ explanatory variables/ attirbutes. 
2- which explanatory variables/ attirbutes are the most significant predictors of the outtput variable.

Loss function:

Loss function is an approach to evaluate how well the algorithm models your dataset. A higher loss is associated with a prediction far from the actual values while a smaller loss function is associated with a prediction closer to the actual values. 

Mean Square Error (MSE) is the most commonly used regression loss function. MSE is the sum of squared distances between our target variable and predicted values.




```python
#graphical
sns.regplot(x="Temperature (C)", y="Apparent Temperature (C)", data=weather_history)
```

```python
from sklearn.linear_model import LinearRegression

#implement linear regression
lm = LinearRegression()
lm.fit(X_train,y_train)

# [1 Mark] - Report Training Data
print('Apperent Teperature = {:.3f} * temperature + {:.3f}'.format(float(lm.coef_), lm.intercept_))
```

Test on test set

```python
#Testing
y_pred = lm.predict(X_test)

# [2 Marks] Plot both Scatters
plt.scatter(X_test, y_test, label='Test Data')
plt.scatter(X_test, y_pred, color='r', marker='+', s=50, label='LR pred.')
```
